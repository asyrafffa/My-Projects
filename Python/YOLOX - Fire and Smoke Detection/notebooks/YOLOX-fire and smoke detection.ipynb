# Train model using custom datasets
### 1. Before training with custom datasets, make sure the dataset have been label. I am using "labelme" to label data. You can read more on their GitHub: https://github.com/labelmeai/labelme
### 2. Split the data into train and validation

```python
import os
import shutil
import random

def split_data(source_folder, train_folder, val_folder, split_ratio=0.7):
    # Create output folders if they don't exist
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(val_folder, exist_ok=True)

    # Get all image files
    image_files = [f for f in os.listdir(source_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.jfif', '.webp'))]
    
    # Shuffle the file list
    random.shuffle(image_files)

    # Calculate split index
    split_index = int(len(image_files) * split_ratio)

    # Split and copy files
    for i, filename in enumerate(image_files):
        src = os.path.join(source_folder, filename)
        if i < split_index:
            dst = os.path.join(train_folder, filename)
        else:
            dst = os.path.join(val_folder, filename)
        shutil.copy2(src, dst)

    # Print statistics
    print(f"Total images: {len(image_files)}")
    print(f"Images in training set: {split_index}")
    print(f"Images in validation set: {len(image_files) - split_index}")

# Set your folders
source_folder = {input directory}
train_folder = os.path.join(os.path.dirname(source_folder), "train")
val_folder = os.path.join(os.path.dirname(source_folder), "val")

# Run the split
split_data(source_folder, train_folder, val_folder)
```

### 3. After labelling the data, convert them into COCO format.

#### Labelme2coco

```python
import json
import os
from collections import defaultdict
from PIL import Image

def labelme_to_coco(labelme_json_dir, output_file):
    # Initialize COCO format structure
    coco_format = {
        "info": {},
        "licenses": [],
        "categories": [],
        "images": [],
        "annotations": []
    }

    # Load all LabelMe JSON files
    labelme_files = [f for f in os.listdir(labelme_json_dir) if f.endswith('.json')]

    # Mapping from category names to IDs
    category_map = {}
    category_id = 1

    image_id = 1
    annotation_id = 1

    for labelme_file in labelme_files:
        labelme_path = os.path.join(labelme_json_dir, labelme_file)
        with open(labelme_path, 'r') as f:
            labelme_data = json.load(f)

        # Extract image information
        image_path = os.path.join(labelme_json_dir, labelme_data['imagePath'])
        img = Image.open(image_path)
        width, height = img.size

        image_info = {
            "id": image_id,
            "file_name": labelme_data['imagePath'],
            "width": width,
            "height": height
        }
        coco_format["images"].append(image_info)

        # Extract annotations
        for shape in labelme_data['shapes']:
            category_name = shape['label']
            if category_name not in category_map:
                category_map[category_name] = category_id
                coco_format["categories"].append({
                    "id": category_id,
                    "name": category_name
                })
                category_id += 1

            category_id_for_shape = category_map[category_name]

            points = shape['points']
            x_coords = [p[0] for p in points]
            y_coords = [p[1] for p in points]
            min_x, max_x = min(x_coords), max(x_coords)
            min_y, max_y = min(y_coords), max(y_coords)

            annotation_info = {
                "id": annotation_id,
                "image_id": image_id,
                "category_id": category_id_for_shape,
                "bbox": [min_x, min_y, max_x - min_x, max_y - min_y],
                "area": (max_x - min_x) * (max_y - min_y),
                "iscrowd": 0
            }
            coco_format["annotations"].append(annotation_info)
            annotation_id += 1

        image_id += 1

    # Save COCO format JSON
    with open(output_file, 'w') as f:
        json.dump(coco_format, f, indent=4)

# Example usage
labelme_json_dir = {input directory}
output_file = {output directory}
labelme_to_coco(labelme_json_dir, output_file)
```

- Under train2017 and val2017 file contain the labelled data
