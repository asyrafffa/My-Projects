## **Extract News Raw Data**

```python
import requests
import pandas as pd

# Set up API credentials and endpoint
API_KEY = 'API_KEY'
API_ENDPOINT = 'API_ENDPOINT'

# Define function to retrieve business articles about Apple stock
def get_business_articles_about_apple_stock(pages=1): #5 pages
    # Set up API parameters
    params = {
        'api-key': 'API_KEY',
        'q': 'Apple',
        'fq': 'news_desk:("Business","Financial","Business Day")',
        'sort': 'newest',
        'page': 0 #change page number here
    }
    articles = []
    for page in range(pages):
        # Send request to API
        response = requests.get(API_ENDPOINT, params=params)
        # Get list of articles from response
        articles += response.json()['response']['docs']
        # Increment page parameter
        params['page'] += 1
    # Convert articles to pandas dataframe
    articles_df = pd.DataFrame(articles)
    return articles_df

# Get business articles about Apple stock from API
df = get_business_articles_about_apple_stock(pages=1) #change pages

# Print dataframe
df.head()
```

## **Preprocessing**

### Feature Selection

```python
# Feature Selection
clean_news_data = df[['abstract', 'pub_date']]
clean_news_data.head()
```

```python
# Making a copy of 'abstract' column to compare sentiment later when applying model
clean_news_data['abstract_2'] = clean_news_data['abstract']
clean_news_data.head()
```

### Change Data Type

```python
# Check data type
print(type(clean_news_data['abstract'].iat[0]))
print(type(clean_news_data['abstract_2'].iat[0]))
print(type(clean_news_data['pub_date'].iat[0]))
```

```python
# Apply datetime to extract date from 'pub_date' column
clean_news_data['date'] = pd.to_datetime(clean_news_data['pub_date']).dt.date
```

```python
# Check data type
print(type(clean_news_data['date'].iat[0]))
```

```python
# Drop 'pub_date' column
clean_news_data = clean_news_data.drop(columns = ['pub_date'])
clean_news_data.head()
```

## **Clean Text**

```python
import nltk
from nltk.stem import WordNetLemmatizer
import re

nltk.download('wordnet')

# init lemmatizer
lemmatizer = WordNetLemmatizer()

from nltk.corpus import stopwords
nltk.download('stopwords')

from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
nltk.download('omw-1.4')
```

```python
# Remove special characters and numbers
clean_news_data['abstract'] = clean_news_data['abstract'].apply(lambda x: re.sub('<.*?>', ' ', re.sub('[^A-Za-z]+', ' ', str(x))))

# Remove short word
clean_news_data['abstract'] = clean_news_data['abstract'].apply(lambda x: ' '.join([word for word in str(x).split() if len(word) > 2]))

# Convert text to lowercase
clean_news_data['abstract'] = clean_news_data['abstract'].apply(lambda x: x.lower())

# Remove stopwords
stop_words = set(stopwords.words('english'))
clean_news_data['abstract'] = clean_news_data['abstract'].apply(lambda x: ' '.join([word for word in word_tokenize(x) if word.lower() not in stop_words]))

# Lemmatize words
# lemmatizer = WordNetLemmatizer()
clean_news_data['abstract'] = clean_news_data['abstract'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(x)]))
```

```python
# Comparing clean and original 'abstract' data
print(clean_news_data['abstract'][0])
print(clean_news_data['abstract_2'][0])
```

## **Apply Sentiment Model**

I am using VADER, one of the models that have been use by many people to extract sentiment from text/news/social media.

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Instantiate SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Apply sentiment analysis to each article's text
sentiment_scores = clean_news_data['abstract'].apply(lambda review:sid.polarity_scores(str(review)))

# Create new columns for each sentiment score
clean_news_data['compound'] = sentiment_scores.apply(lambda x: x['compound'])

# Apply sentiment analysis to each article's text
sentiment_scores = clean_news_data['abstract_2'].apply(lambda review:sid.polarity_scores(str(review)))

# Create new columns for each sentiment score for unclean data
clean_news_data['compound_2'] = sentiment_scores.apply(lambda x: x['compound'])

# Print dataframe
clean_news_data.head()
```

### Extract Label

```python
# Function to label sentiment
def label_sentiment(score):
    if score < 0:
        return 'negative'
    elif score == 0:
        return 'neutral'
    else:
        return 'positive'

# Labeling sentiment for both columns
clean_news_data['sentiment_label_1'] = clean_news_data['compound'].apply(label_sentiment)
clean_news_data['sentiment_label_2'] = clean_news_data['compound_2'].apply(label_sentiment)

clean_news_data.head()
```

## **Visualisation**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Melt DataFrame for easier plotting
melted_data = clean_news_data.melt(value_vars=['sentiment_label_1', 'sentiment_label_2'],
                                     var_name='Score_Type', value_name='Sentiment')

# Count occurrences of each sentiment
sentiment_counts = melted_data.groupby(['Score_Type', 'Sentiment']).size().reset_index(name='Count')

# Side-by-side bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=sentiment_counts, x='Sentiment', y='Count', hue='Score_Type', palette='pastel')
plt.title('Comparison of Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.legend(title='Score Type')
plt.show()
```

```python
# Aggregate by date (mean sentiment score)
daily_sentiment = clean_news_data.groupby('date')[['compound', 'compound_2']].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_sentiment, x='date', y='compound', marker='o', label='Text 1 Sentiment')
sns.lineplot(data=daily_sentiment, x='date', y='compound_2', marker='o', label='Text 2 Sentiment')

plt.title('Average Sentiment Score Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sentiment Score')
plt.axhline(0, color='gray', linestyle='--')  # Neutral sentiment line
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Melt DataFrame for easier plotting
melted_data = clean_news_data.melt(id_vars='date', value_vars=['sentiment_label_1', 'sentiment_label_2'],
                       var_name='Sentiment_Type', value_name='Sentiment')

# Count occurrences of each sentiment (excluding neutral)
sentiment_counts = melted_data[melted_data['Sentiment'] != 'neutral'] \
    .groupby(['date', 'Sentiment']).size().reset_index(name='Count')

# Pivot for plotting
pivot_counts = sentiment_counts.pivot_table(index='date', columns='Sentiment', values='Count', fill_value=0)

# Create a horizontal bar chart
plt.figure(figsize=(12, 6))
# Plot positive as is, and negative as negative values
plt.barh(pivot_counts.index, pivot_counts['positive'], color='green', label='Positive')
plt.barh(pivot_counts.index, -pivot_counts['negative'], color='red', label='Negative')

plt.title('Sentiment Distribution Over Time (Excluding Neutral)')
plt.xlabel('Count')
plt.ylabel('Date')
plt.axvline(0, color='black', linewidth=0.8)  # Add a vertical line at x=0
plt.legend(title='Sentiment')
plt.grid(axis='x')
plt.show()
```

```python
# Compare sentiments and create a new column indicating if they differ
clean_news_data['sentiment_diff'] = clean_news_data['sentiment_label_1'] != clean_news_data['sentiment_label_2']

# Filter rows where sentiments differ
different_sentiments = clean_news_data[clean_news_data['sentiment_diff']]

different_sentiments.head()
```
